```python
import sys
import os
sys.path.append('..')

from pymir import AudioFile
from pymir import Energy
from pymir import Onsets

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
%pylab inline
```

```python
def get_segments(file_genre, file_name, num_segments = 300):
    filepath = os.path.join("../", "data", file_genre,file_name)
#     print filepath
    wav_data = AudioFile.open(filepath)
    segments = wav_data.frames(len(wav_data)/num_segments)
    return segments

segments1 = get_segments("disco","disco.00000.wav")
segments2 = get_segments("disco","disco.00001.wav")
segments3 = get_segments("disco","disco.00002.wav")
```

```python
centroids1 = [f.spectrum().centroid() for f in segments1]
centroids2 = [f.spectrum().centroid() for f in segments2]
centroids3 = [f.spectrum().centroid() for f in segments3]
```

```python
plt.plot(centroids1, color='blue',hold=True)
plt.plot(centroids2, color='red', hold=True)
plt.plot(centroids3, color='green', hold=True)
plt.show()
```

```python
chroma_feature = 0
chroma1 = [f.spectrum().chroma()[chroma_feature] for f in segments1]
chroma2 = [f.spectrum().chroma()[chroma_feature] for f in segments2]
chroma3 = [f.spectrum().chroma()[chroma_feature] for f in segments3]
```

```python
zcr1 = [f.zcr() for f in segments1]
zcr2 = [f.zcr() for f in segments2]
zcr3 = [f.zcr() for f in segments3]
```

```python
plt.plot(zcr1, color='blue',hold=True)
plt.plot(zcr2, color='red', hold=True)
plt.plot(zcr3, color='green', hold=True)
plt.show()
```
